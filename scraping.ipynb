Scrape the content of https://www.lemonde.fr/Links to an external site. and save it as a CSV.

We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

Tips:

You're going to be missing some article URLs - you should use try/except on them to avoid them being a problem, buuuut after that I would recommend printing the element to see why your approach isn't working (it's 100% possible to get all of the URLs!)
If you want to make sure the element you're getting has a specific attribute, you can do something like .select_one("a[href]"). That will only give you anchor tags (links) that have hrefs.
Instead of yes/no for the premium question, you can think of it as "put text in this column if the article is premium, don't put text in it if it is not"

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.lemonde.fr/")
doc = BeautifulSoup(response.text)

print(doc.prettify())
```

```python
# We want: titles, subhead, article URL, whether it's premium or not, byline, article type, image URL.

articles = doc.find_all(class_="article")
urls_seen = set()


len(articles)
print(f"My selection contains {len(articles)} articles.")

articles_list = []

for item in articles:
    # Title
    title = item.find(class_= 'article__title').text

    # Premium?
    premium_tag = item.find(class_ = 'icon__premium')
    if premium_tag:
        premium = 'is prenium'
    else:
        premium = None

    # byline?
    author = item.find(class_ = 'article__author-name')
    if author:
        author = author.text
    else:
        author = None

    # article type?
    article_cat = item.find(class_ = 'article__type')
    if article_cat:
        article_type = article_cat.text
    else:
        article_type = None
    
    # subhead?
    subhead_tag = item.find(class_ = 'article__desc')
    if subhead_tag:
        subhead = subhead_tag.text
    else:
        subhead = None
    
    # image URL
    img_tag = item.find('img')
    if img_tag:
        img_url = img_tag.get('src') or img_tag.get('data-src')
    else:
        img_url = None

    # URL
    # url_tag = item.find('a')
    # if url_tag:
    #     url = url_tag['href']
    # else:
    #     url_other = item.find('div', class_='article')
    #     if url_other:
    #         url = url_other.find('a')['href']

    url_tag = item.find('a', href=True)
    url = url_tag['href'] if url_tag else None

    # Sauter si url déjà vue
    if not url or url in urls_seen:
        continue
    urls_seen.add(url)

    # # Print all
    # print("------")
    # print(title)
    # print(url)
    # print(article_type)
    # print(author)
    # print(img_url)

    articles_list.append({
    'title': title,
    'subhead': subhead,
    'author': author,
    'url': url,
    'premium': premium,
    'article_type': article_type,
    'img_url': img_url
    })

df_articles = pd.DataFrame(articles_list)
df_articles
```

```python
df_articles.to_csv('le_monde_articles.csv')
```

# Part Three

Scrape the list of third party drivers license locations from https://travel-id-documents.az.gov/authorized-third-party-driver-license-locationsLinks to an external site. but include the link. Save as a CSV. Since it's more than just the text from the table, this requires actually using BeautifulSoup :(

Tips:

You'll need user-agent headers for this one
In class we always did something like item.find('h2') or the like because there was only ever one h2 we wanted. In this case there are four <td> tags that you want the text from! You'll need to use .find_all and treat them like normal lists ("give me the text from the first one," + "give me the text from the second one" etc etc)
It's okay to have columns you don't want or need! You can always remove them later

```python


import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac macOS 10_15_7) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}
response = requests.get("https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations", headers=headers)
doc = BeautifulSoup(response.text)

print(doc.prettify())
```

```python
tds = doc.find_all('td')

# Regrouper les <td> par blocs de 4
rows = [tds[i:i+4] for i in range(0, len(tds), 4)]

# Extraire le texte de chaque cellule dans chaque ligne
table_data = []
for row in rows:
    # on vérifie que la ligne contient bien 4 éléments
    if len(row) == 4:
        table_data.append([td.get_text(strip=True) for td in row])

# Résultat : une liste de listes (comme un tableau 2D)
for row in table_data:
    print(row)

header = table_data[0]
rows = table_data[1:]
df = pd.DataFrame(rows, columns=header)
df.to_csv('third_party_driver_license.csv', index=False)
```

Visit https://www.tnwb.uscourts.gov/Search/Search.aspx Links to an external site. and search for "CAR." Scrape the results into a CSV, with four columns: the URL to the case, the name of the case, the category (e.g. "Judge's Opinions), the additional details (terms match/size/pdf URL).

```python
import requests
from bs4 import BeautifulSoup


# "https://www.tnwb.uscourts.gov/Search/Search.aspx?zoom_sort=0&zoom_xml=0&zoom_query=CAR&zoom_per_page=10&zoom_and=1&zoom_cat%5B%5D=-1"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac macOS 10_15_7) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36'}
response = requests.get("https://www.tnwb.uscourts.gov/Search/Search.aspx?zoom_sort=0&zoom_xml=0&zoom_query=CAR&zoom_per_page=200&zoom_and=1&zoom_cat%5B%5D=-1", headers=headers)
doc = BeautifulSoup(response.text)

print(doc.prettify())
```

```python
# the URL to the case, the name of the case, the category (e.g. "Judge's Opinions), the additional details

# All the case:
# class="result_altblock"

# ----
# URL to case == a href="https://www.tnwb.uscourts.gov/Opinions/ghb/pdf/ghb19981105xn1.pdf#search=%22car%22" 
# Name to case == <a href="https://www.tnwb.uscourts.gov(...))" target="_blank">GHB: 98-12504 Ricky D. Alston and Penny M. Alston</a> 
# Cat of Jugement == class="category" 
# Additional details == class="infoline" 

items = doc.find_all('div', class_='result_altblock')

ls_cases = []

for item in items:
    # Name and URL
    a_tag = item.find("a")
    if a_tag:
        case_url = a_tag.get("href")
        case_name = a_tag.get_text(strip=True)
    else:
        case_url = case_name = None

    # Category
    category_tag = item.find("span", class_="category")
    category = category_tag.get_text(strip=True) if category_tag else None

    # Details
    infoline_tag = item.find("div", class_="infoline")
    details = infoline_tag.get_text(strip=True) if infoline_tag else None

    # Affichage ou stockage
    # print({
    #     "case_url": case_url,
    #     "case_name": case_name,
    #     "category": category,
    #     "details": details
    # })

    ls_cases.append({
        "case_url": case_url,
        "case_name": case_name,
        "category": category,
        "details": details
    })

df_cases = pd.json_normalize(ls_cases)
df_cases
```

```python
df_cases.to_csv("cases.csv", index=False)
```
